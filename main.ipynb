```python
# count the ids in the img_df that are also in the english_df
print(img_df['id'].isin(english_df['id']).sum())

# count the ids in the img_df that are also in the tab_df
print(img_df['id'].isin(tab_df['id']).sum())

# Create a boolean mask for entries that exist in all three datasets
common_ids_mask = img_df['id'].isin(english_df['id']) & img_df['id'].isin(tab_df['id'])

# Get the common IDs
common_ids = img_df[common_ids_mask]['id']

# Display entries from all three datasets that have matching IDs
display(img_df[img_df['id'].isin(common_ids)])
display(english_df[english_df['id'].isin(common_ids)])
display(tab_df[tab_df['id'].isin(common_ids)])
```
